### 1. Basic Statistics

Use Python to read in your data file and show many rows and columns your data has.

```python

import pandas as pd

data = pd.read_csv('/content/Heart_Disease_Prediction.csv')

print(f'Total rows: {data.shape[0]}')
print(f'Total columns: {data.shape[1]}')
```

### 2. Data Diversity

Use Python to identify how many unique values each column in your data has.  Use Python to identify which column has the greatest number of distinct values (aka the most diversity)?  If your data has a unique identifier (e.g. a person or visit or record ID) then find the column with the next most number of unique values.

```python

unique_counts = data.nunique()

most_diverse = unique_counts.idxmax()
most_diverse_count = unique_counts.max()

print('Unique value counts per column:')
print(unique_counts)
print(f'Column with the most diversity: {most_diverse} ({most_diverse_count} unique values)')
```

### 3. Data Redundancy

Use Python to identify any columns in your data where the value for every row is the same.  For example, if you had a gender column and all the rows said 'M', then you would want to write the code that helped you determine that.

```python
import pandas as pd


df = pd.read_csv('Heart_Disease_Prediction.csv')


redundant_columns = [col for col in df.columns if df[col].nunique() == 1]

print('Columns with redundant values:')
print(redundant_columns)
```

### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/content/Heart_Disease_Prediction.csv')

numeric_column = 'Age'


plt.figure(figsize=(8, 6))
plt.hist(df[numeric_column], bins=20, color='skyblue', edgecolor='black')
plt.title(f'Distribution of {numeric_column}', fontsize=14)
plt.xlabel(numeric_column, fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('Heart_Disease_Prediction.csv')

col_x = 'Age'
col_y = 'Cholesterol'


plt.figure(figsize=(8, 6))
sns.regplot(data=df, x=col_x, y=col_y, scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})
plt.title(f'Relationship between {col_x} and {col_y}', fontsize=14)
plt.xlabel(col_x, fontsize=12)
plt.ylabel(col_y, fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```
